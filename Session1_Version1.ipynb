# Problem 1: Hunny Hunt

```python
def linear_search(items, target):
    for i in range(len(items)):
        if items[i] == target:
            return i
    return -1

items = ['haycorn', 'haycorn', 'haycorn', 'hunny', 'haycorn']
target = 'hunny'
x = linear_search(items, target)
print(x)
```

```python
items = ['bed', 'blue jacket', 'red shirt', 'hunny']
target = 'red balloon'
print(linear_search(items, target))
```

# Problem 2: Bouncy, Flouncy, Trouncy, Pouncy

```python
def final_value_after_operations(operations):
    tigger = 1
    for ele in operations:
        if (ele == 'bouncy') or (ele == "flouncy"):
            tigger += 1
        elif (ele == "trouncy") or (ele == "pouncy"):
            tigger -= 1
    return tigger


operations = ["trouncy", "flouncy", "flouncy"]
print(final_value_after_operations(operations))

operations = ["bouncy", "bouncy", "flouncy"]
print(final_value_after_operations(operations))
```

# Problem 3: T-I-Double Guh-Er II

```python
def tiggerfy(word):
    word = word.lower()
    substring = ['t', 'i', 'gg', 'er']
    for i in substring:
        word = word.replace(i, "")
    return word
            
        
word = "Trigger"
print(tiggerfy(word))

word = "eggplant"
print(tiggerfy(word))

word = "Choir"
print(tiggerfy(word))
```
